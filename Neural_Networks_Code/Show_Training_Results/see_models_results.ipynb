### MOUNT AND IMPORT

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

plt.style.use("ggplot")
%matplotlib inline
```

```python
# Path where history is stored
history_path = 'C:\\Users\\Usuario\\Desktop\\ultimate\\h\\'
all_hist = os.listdir(history_path)

# For all history of the models
for hist in all_hist:
    
    # Print results 
    results = pd.read_csv(os.path.join(history_path,hist));
    print(results['val_accuracy'][len(results['val_accuracy'])-1])
    try:
        print(results['val_mean_io_u_1'][len(results['val_mean_io_u_1'])-1])
    except:
        try:
            print(results['val_mean_io_u'][len(results['val_mean_io_u'])-1])
        except:
            print(results['val_mean_io_u_2'][len(results['val_mean_io_u_2'])-1])

    print(results['val_f1_score'][len(results['val_f1_score'])-1])
    print(results['val_precision_m'][len(results['val_precision_m'])-1])
    print(results['val_recall_m'][len(results['val_recall_m'])-1])
    print(results['val_dice_loss'][len(results['val_dice_loss'])-1])
    print()
    
    # Loss curve
    fig, ax = plt.subplots(figsize=(8,5))
    ax.plot(results["loss"], label="loss")
    ax.set_title("Learning curve " + hist)
    ax.plot(results["val_loss"], label="val_loss")
    ax.plot( np.argmin(results["val_loss"]), np.min(results["val_loss"]), marker="x", color="r", label="best model")
    print(np.argmin(results["val_loss"]), np.min(results["val_loss"]))
    
    ax.set_xlabel("Epochs")
    ax.set_ylabel("log_loss")
    ax.legend();

    fig.tight_layout()
    plt.show()

    # Accuracy curve
    fig, ax = plt.subplots(figsize=(8,5))
    ax.plot(results["accuracy"], label="dice_loss")
    ax.set_title("Accuracy curve " + hist)
    ax.plot(results["val_accuracy"], label="val_dice_loss")
    #ax.plot( np.argmin(results["dice_loss"]), np.min(results["dice_loss"]), marker="x", color="r", label="best model")
    ax.set_xlabel("Epochs")
    ax.set_ylabel("Accuracy (dice-loss)")
    ax.legend();

    fig.tight_layout()
    plt.show()
    
```
