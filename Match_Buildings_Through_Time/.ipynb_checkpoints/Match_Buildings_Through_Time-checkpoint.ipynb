## Cargar todas las imágenes y obtener todos los polígonos de los edificios 

```python
imgs_pred = df_folder
geoms_list = [] # Guarda los polígonos

# Se establece una área mínima y un threshold para obtener los polígonos de una imagen 
min_area = 3.5 # in pixels
bg_threshold = 0
simplify = False
print("bg_threshold:", bg_threshold)
print("min_area:", min_area)

# Leer cada imagen de imgs_pred y obtener polígonos de esa imagen a partir de la máscara predicha (mask_image)
# Guardar todos los polígonos de cada imagen en geoms_list
for name_img in tqdm(imgs_pred):
    mask_image = skimage.io.imread(name_img)
    geoms = sol.vector.mask.mask_to_poly_geojson(mask_image, 
                                                 min_area=min_area,
                                                 bg_threshold=bg_threshold,
                                                 simplify=simplify)
    
    
    geoms_list.append(geoms)
    #print("N geoms:", len(geoms))   
```

## Emparejamiento de edificios 

### Parte 1: Emparejamiento de todos los edificios con todos 

```python
# Guardar el área de todos los polígonos
all_areas = []
all_areas_value = []

for i in range(len(geoms_list)):
  all_areas.append(geoms_list[i]['geometry'].area)
  all_areas_value.append(list(all_areas[i].values))
```

```python

lista_i = []
lista_a = []

# Para cada conjunto de polígonos (edificios) de cada foto 
for i in tqdm(range(len(geoms_list)-1)):
    lista_i.append([])
    lista_a.append([])
    print("---")
    print("nuevo folder " + str(i)) 
    # Se obtienen los polígonos de otra foto posterior distinta de la actual 
    #(se compara los polígonos de la foto 1 con la 2,1-3,1-4,..., 2-3,2-4,...3-4,3-5) ()
    for i2 in (range(i+1, len(geoms_list))):
        print(i)
        print(i2)
        lista_i[i].append([])
        lista_a[i].append([]) 
        id2 = i2-i-1
        
        # Inicializo los arrays a -1 para todos los polígonos
        # El parecido con cada polígono (de i) con los polígonos de i2, se asocian con los índices de lista_i y lista_a
        for n in (range(len(geoms_list[i]))):  # guardo n arrays del tamaño que estoy viendo y cada uno corresponde con el idx
            lista_i[i][id2].append(-1)
            lista_a[i][id2].append(-1)
        
        # Para cada polígono de la imagen de i2
        for j in (range(len(geoms_list[i2]['geometry']))):
                # Obtener área de solape entre el polígono j de i2 y todos los polígonos de i
                value_percent = geoms_list[i]['geometry'].intersection(geoms_list[i2]['geometry'][j]).area / all_areas_value[i]
                
                # Si hay un solapamiento mayor al 10% entre algún polígono de la imagen i 
                # con el polígono j de la imagen i2 actual 
                if(any(value_percent) > 0.1):
                  # Obtener solapamiento máximo 
                  value_percent_list = list(value_percent.values)
                  maxim = max(value_percent)
                  idx = value_percent_list.index(maxim)
                  
                  # Guardar el índice correspondiente a este solapamiento y el área
                  # La correspondencia j (polígono imagen i2) con el polígono idx (polígono imagen i)
                  # se guarda según los índices LISTA[i][i2][idx] = j
                  # Correspondencia lista_i[imagen1][imagen2][polígono_actual] = polígono_correspondiente_de_i2
                  # Lista del área solapada: lista_a[imagen1][imagen2][polígono_actual] = área_solapada
                    
                  # Si ya tenía alguna correspondencia, si el área solapada es mayor, se sustituye
                  if(lista_i[i][id2][idx] > -1):  
                        if(lista_a[i][id2][idx] < maxim):
                            lista_i[i][id2][idx] = j
                            lista_a[i][id2][idx] = maxim
                  else:
                       lista_i[i][id2][idx] = j
                       lista_a[i][id2][idx] = maxim 
```

```python
# Obtener los valores en forma de lista de números y eliminar nan
df_val = []
for g in tqdm(range(len(dfit))):
    df_val.append([])
    for h in (range(len(dfit[g]))):
      if( str(dfit[g][h])!= 'nan'):
        df_val[g].append(eval(dfit[g][h]))
```

```python
# Comprobar si el polígono actual ya se había añadido a una lista anterior de concatenaciones o es que ha comenzado una nueva
# lista de polígonos que no incluían los anteriores 
def checkPrev (h, idx,j):
    prev = False
    breaked = False
    if(h > 0):
        for f in range(h):
            if(breaked):
                break
            for c in range(len(concat_l[f])):
                if(len(concat_l[f][c]) > (h-f) ):
                    if(concat_l[f][c][h-f] == j):
                        prev = True
                        breaked = True
                        break
    return prev
```

```python
#%%capture

# Se desea crear una lista con la concatenación de todos los polígonos que tienen correspondencia entre sí de forma continuada
# En este bucle, se crea solo para cada imagen con sus imágenes inmediatamente posteriores
# En este caso, lo que ocurre es lo siguiente: 
# - Si un polígono (j) de la imagen 0, tiene correspondencia con un polígono de la imagen 1 posterior (idx), 
# se guarda en una lista, j e idx
# - Si este polígono idx tiene correspondencia con algún polígono de la imagen posterior (en este caso imagen 2) (nuevo idx),
# se guarda esta correspondencia a la lista anterior
# - Se sigue avanzando para todos los polígonos de todas las imágenes. Si no hay correspondencia, se dejan de añadir polígonos. 
# - Se comienza ahora por la imagen h=1 y se ve las correspondencias de sus polígonos con la imagen h=2 y las correspondencias 
# posteriores. 
# - Si alguna correspondencia de la imagen nueva h=1, ya estaba en una lista anterior de h=0, no se hace una nueva lista, esto 
# se comprueba con checkPrev

# Para toda la lista de polígonos correspondientes guardados en df_val
concat_l = []
for h in tqdm(range(len(df_val))):
    concat_l.append([])
    
    # Para todos los polígonos de la imagen h en correspondencia con la imagen posterior inmediata [0]
    # En el caso la imagen h=0, las correspondencias en [0], serán de la imagen h=1
    # Para la imagen h=1, las corrspondencias en [0], serán de la imagen h=2, etc
    for j in (range(len(df_val[h][0]))):
        concat_l[h].append([])
        concat_l[h][j].append(j)  # Guardar polígono actual como correspondencia 
        idx = df_val[h][0][j] # Polígono correspondiente
        
        # Si el polígono [h][0][j] ([imagen_h][imagen_posterior_inmediata]) tiene correspondencia (!=-1)
        if(idx > -1):
            # Si el polígono actual no estaba en una lista anteiror
            if(not checkPrev(h, idx,j)):
                concat_l[h][j].append(idx) # Guardar polígono correspondiente 
                
                # Se comprueba las correspondencias de los polígonos posteriores y se añaden o se termina de buscar
                for n in range(h+1,len(df_val)):
                    idx = df_val[n][0][idx]

                    if(idx > -1):
                        concat_l[h][j].append(idx)
                    else:
                        break    
```

```python
ultimo_len = len(geoms_list[-1]['geometry'])
ultimo_len
```

```python
# En esta parte se añaden -1 en las listas anteriores cuando no hay emparejamientos con polígonos en imágenes intermedias
# y se añaden los id de los polígonos que si tienen emparejamientos en imágenes posteriores. 
# Por lo tanto, antes teníamos una lista con los id de polígonos que era [0,15,33,2], que indica que para el polígono 0,
# de la imagen 0, tiene correspondencia con el polígono 15, de la imagen 1, el 33 de la imagen 3,... y a partir de la imagen 3, 
# ya no existen más correspondencias. Ahora se comprueba si hay más correspondencias a partir de estas. Entonces la imagen 3
# se compara con la imagen 5,imagen 6,... añadiendo las correspondencias pertinentes y los saltos en medio: 
# [0, 15, 33, 2,-1,-1, 5, -1, 40]. 
# Esto indica que ahora el polígono actual no tiene correspondencia con la imagen 4 ni 5 pero si tiene en la 6 y 7 y 9.

# Lista que guarda los id concatenados
concat_j2 = []

# Para cada imagen
for h in tqdm(range(0,len(concat_l))):         # para cada foto
    concat_j2.append([])
    
    # Para cada polígono de la imagen 
    for j in (range(len(concat_l[h]))):  
        concat_j2[h].append([])
        concat_j2[h][j].append(j)
        
        # Hasta completar los 23,22,21... matches
        # (No se repiten matches 1-2,2-1,...por eso voy restando h)
        for l in range(len(concat_l)-h):   
            # Si el polígono correspondiente j no se encuentra en las listas de correspondencia de la imagen h 
            if(not checkIn(h,j)):
                # Obtiene la siguiente imagen (final_i) en la que hay que mirar si hay correspondencia 
                # de los polígonos de concat_j2[h][j] 
                final_i = len(concat_j2[h][j]) - 1 + h  # Importante la h
                idx = concat_j2[h][j][-1] # Correspondencia final 
                
                # Si ya estamos al final de las imágenes (tam_1==total imágenes) se termina de buscar 
                if(final_i == tam_1):
                    break
                    
                # Para cada polígono de la imagen final_i, se comprueba si tiene correspondencias 
                # con polígonos en imágenes posteriores
                for n in range(0,len(df_val[final_i])):  
                    value = df_val[final_i][n][idx]  # Obtener polígono correspondiente 
                    if(value > -1): # Si hay correspondencia 
                        # Si no se ha añadido a otra lista anterior
                        if(not checkIn(h,value)):
                            # Añadir tantos -1 a la lista como imágenes intermedias que no tienen 
                            # según las imagenes intermedias que no tienen correspodencia en la lista 
                            for add in range(final_i, final_i+n):
                                concat_j2[h][j].append(-1)
                                
                            # Añadir correspondencia a la lista 
                            index = df_val[final_i][n][idx]
                            concat_j2[h][j].append(index)
                            break;
```
